```python
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
billboard = pd.read_csv('billboard.csv')

mask_10 = (billboard['weeks_on_chart'] >= 10)#.drop_duplicates()

songs_10 = billboard[mask_10]

#songs_10.drop_duplicates(subset='song_id')

by_artist = songs_10.groupby('performer')['song'].nunique()
artist_30 = by_artist[by_artist >= 30]
#print(len(artist_30))
#type(artist_30)
```

```python
plt.figure(figsize=(10, 6))
artist_30.plot(kind='bar')
plt.xticks(rotation=90)
plt.xlabel('Performers')
plt.ylabel('Number of Songs')
plt.title('Number of Songs by Performer')
plt.tight_layout()
plt.show()
```
