```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from the CSV file
data = pd.read_csv("capmetro_UT.csv")

# Create a facet grid plot
g = sns.FacetGrid(data, col="day_of_week", col_wrap=3, height=4)

# Plot boarding counts
g.map(sns.lineplot, "hour_of_day", "boarding", ci=None, color='blue')

# Plot temperature
g.map(sns.lineplot, "hour_of_day", "temperature", ci=None, color='red')

# Set axis labels and titles
g.set_axis_labels("Hour of Day", "Count/ Temperature in degrees F")
g.set_titles(col_template="{col_name}")
plt.subplots_adjust(top=0.9)

# Add legend manually
boarding_line = plt.Line2D([], [], color='blue', label='Boarding')
temperature_line = plt.Line2D([], [], color='red', label='Temperature')
plt.legend(handles=[boarding_line, temperature_line], loc='best')

# Add a figure caption in paragraph form
caption_lines = [
    'This figure shows use the trends in boarding at different times of day over each day of the week.',
    'In addition, we also tracked daily temperature over time over each day of the week to see if there was an relationship',
    'between higher temperatures and people choosing to get on the bus.',
    'We generally saw this trend, in addition to seeing that more people tend to board the bus in the middle of the day',
    'on weekdays since those are the days people work or have class.'
]

caption_y = 0.00  # Starting y-coordinate for the caption
line_height = 0.03  # Height between each line

for line in caption_lines:
    plt.figtext(0.1, caption_y, line, ha='left', fontsize=10, color='gray')
    caption_y -= line_height
g.fig.suptitle("Capital Metro Ridership and Temperature Patterns by Day of Week")
plt.show()
```
